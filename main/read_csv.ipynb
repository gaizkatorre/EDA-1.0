## Limpieza del Dataset

```python
df_vehiculos_limpio = df_partial.drop(['url', 'region_url', 'image_url', 'county', 'lat', 'long', 'region', 'VIN', 'description', 'id', 'posting_date', 'size'], axis=1)
```

```python
for i in df_vehiculos_limpio.columns:
    print(i,len(df_vehiculos_limpio[i].value_counts()))
```

```python
df = df_vehiculos_limpio
```

```python
nan_value = float("NaN")
df["price"].replace(0, nan_value, inplace = True)
df["price"].replace(1, nan_value, inplace = True)

df.dropna(subset = ["price"], inplace=True)
```

```python
df.loc[df['price'].idxmax()]
```

```python
condition_counts = df['condition'].value_counts()
condition_counts
```

```python
df.condition.unique()
```

```python
df['condition'].replace(np.nan, 'fair', inplace=True)
```

```python
df['fuel'].replace(np.nan, 'gas', inplace = True)
```

```python
df.describe()
```

```python
df['odometer'] = pd.to_numeric(df['odometer'], errors='coerce')
```

```python
low_odometer= df[(df['odometer'] < 500)].index
high_odometer = df[df['odometer'] > 300000].index
```

```python
df.drop(index=low_odometer, axis=0, errors="ignore", inplace=True)
df.drop(index=high_odometer, axis=0, errors="ignore", inplace=True)
```

```python
#Cambiar milas a km

df['odometer'] = df['odometer'] * 1.60934
```
